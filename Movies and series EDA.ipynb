# Recommendation system 

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Prepare the data
df_recommend = df.copy()

# Fill missing genres with empty strings
df_recommend['Genre'] = df_recommend['Genre'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))
df_recommend['Genre'] = df_recommend['Genre'].fillna('')

# Combine features into a single string
df_recommend['features'] = (
    df_recommend['Genre'] + ' ' +
    df_recommend['Type'] + ' ' +
    df_recommend['Hidden Gem Score'].astype(str) + ' ' +
    df_recommend['IMDb Score'].astype(str)
)

# Step 2: Vectorize the text using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_recommend['features'])

# Step 3: Compute similarity scores
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Step 4: Build a recommendation function
indices = pd.Series(df_recommend.index, index=df_recommend['Title']).drop_duplicates()

def recommend(title, num_recommendations=5):
    idx = indices.get(title)

    if idx is None:
        return "Title not found in dataset."

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]

    movie_indices = [i[0] for i in sim_scores]
    return df_recommend[['Title', 'Genre', 'IMDb Score']].iloc[movie_indices]

```

## recommendation examples

```python
recommend("Avengers Assemble",5)
```

```python
recommend("The Godfather",5)
```
